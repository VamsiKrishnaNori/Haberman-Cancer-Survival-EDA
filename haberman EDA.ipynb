Exploratary Data Analysis On Haberman Cancer Survival Dataset

```python
#importing required modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
haber=pd.read_csv("haberman.csv")
```

```python
#number of points
print(haber.shape)
```

1.There are total of 306 data points in the data set.

```python
#features
print(haber.columns)
```

2.There are total 4 features they are Age,year_Of_Operation,Positive_Auxilary_Nodes which are independent and Status_of_Survival which is dependent.

3.There are total of two classes calss 1-Survived more than 5 years after operation, class 2-died in less than 5 years of operation.

```python
#classes and data points per feature
haber['status'].value_counts()
```

4.Total there are 225  data points in class 1 and 81 data points in class 2

```python
#scatter plot
#1 scatter plot for age vs positive auxilary nodes
sns.set_style("whitegrid")
sns.FacetGrid(haber,hue="status",size=4)\
   .map(plt.scatter,"age","positive")\
    .add_legend()
plt.title('Scatter plot for Age vs Positive Auxilary Nodes')
plt.xlabel('Age')
plt.ylabel('Positive Auxilary Nodes')
plt.show()
#2 scatter plot for age vs year of operation
sns.set_style("whitegrid")
sns.FacetGrid(haber,hue="status",size=4)\
   .map(plt.scatter,"age","year")\
    .add_legend()
plt.title('Scatter plot for Age vs Year of Operation')  
plt.xlabel('Age')
plt.ylabel('Year of Operation')
plt.show()
#3 scatter plot for year of operation vs positive auxilary nodes
sns.set_style("whitegrid")
sns.FacetGrid(haber,hue="status",size=4)\
   .map(plt.scatter,"year","positive")\
    .add_legend()
plt.title('Scatter plot for Year of Operation vs Positive auxilary nodes')  
plt.xlabel('Year of Operation')
plt.ylabel('Positive Auxilary Nodes')
plt.show()
```

```python
#pairplots
sns.set_style("whitegrid")
sns.pairplot(haber,hue='status',vars=['age','year','positive'],size=3)
plt.show()
```

5.As we have 3 independent features we will get 3c2=3 pair plots that are between Age,Year_of_operation,Positive_auxilary_nodes.

6.In scatter Plots and pair plots we can see that the points are not well distinguished from each classes Which means that we can not get a good classifier that can distinguish between the Survived and dead.

```python
haberman_survive=haber.loc[haber["status"]==1]
haberman_died=haber.loc[haber["status"]==2]
```

```python
#1d scatter plot for age
plt.plot(haberman_survive["age"],np.zeros_like(haberman_survive['age']))
plt.plot(haberman_died["age"],np.zeros_like(haberman_died['age']))
plt.xlabel('Age')
plt.title('1d scatter plot Age')
plt.show()
#1d scatter plot for year of operation
plt.plot(haberman_survive["year"],np.zeros_like(haberman_survive['year']))
plt.plot(haberman_died["year"],np.zeros_like(haberman_died['year']))
plt.title('1d scatter plot Year of Operation')
plt.xlabel('Year Of Operation')
plt.show()
#1d scatter plot for positive auxilary nodes
plt.plot(haberman_survive["positive"],np.zeros_like(haberman_survive['positive']))
plt.plot(haberman_died["positive"],np.zeros_like(haberman_died['positive']))
plt.title('1d scatter plot Positive Auxilary Nodes')
plt.xlabel('Positive Auxilary Nodes')
plt.show()
```

```python
#DISTRIBUTION PLOTS
#distribution plot for age
sns.FacetGrid(haber,hue='status',size=5)\
   .map(sns.distplot,"age")\
    .add_legend()  
plt.title('Distribution Plot of Age')
plt.xlabel('Age')
plt.ylabel('Range')
plt.show()        
#distribution plot for year of operation
sns.FacetGrid(haber,hue='status',size=5)\
   .map(sns.distplot,"year")\
    .add_legend()
plt.title('Distribution Plot of Year of Operation')
plt.xlabel('Year of Operation')
plt.ylabel('Range')
plt.show()
#distribution plot for positive auxilary nodes
sns.FacetGrid(haber,hue='status',size=5)\
   .map(sns.distplot,"positive")\
    .add_legend()
plt.title('Distribution Plot for Positve Auxilary Nodes')
plt.xlabel('Positve Auxilary Nodes')
plt.ylabel('Range')
plt.show()
```

7.By Doing Distribution plot we have known that most of the survivors have 0 positive_auxilary_nodes.

```python
#PDF and CDF for haberman survivors
counts,bin_edges=np.histogram(haberman_survive['age'],bins=10,density=True)
pdf=counts/(sum(counts))
print("1.PDF and CDF of Haberman dataset Survived on Age")
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('PDF and CDF of Haberman dataset Survived on Age')
plt.xlabel('Age')
plt.ylabel('Probability')
plt.show()
counts,bin_edges=np.histogram(haberman_survive['year'],bins=10,density=True)
pdf=counts/(sum(counts))
print("2.PDF and CDF of Haberman dataset Survived on Year of operation")
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('PDF and CDF of Haberman dataset Survived on Year of Operation')
plt.xlabel('Year of Operation')
plt.ylabel('Probability')
plt.show()
counts,bin_edges=np.histogram(haberman_survive['positive'],bins=10,density=True)
pdf=counts/(sum(counts))
print("3.PDF and CDF of Haberman dataset Survived on Positive Auxilary Nodes")
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('PDF and CDF of Haberman dataset Survived on Positive Auxilary Nodes')
plt.xlabel('Positive Auxilary Nodes')
plt.ylabel('Probability')
plt.show()
```

```python
#PDF and CDF for haberman died
counts,bin_edges=np.histogram(haberman_died['age'],bins=10,density=True)
pdf=counts/(sum(counts))
print("1.PDF and CDF of Haberman dataset died on Age")
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('PDF and CDF of Haberman dataset Died on Age')
plt.xlabel('Age')
plt.ylabel('Probability')
plt.show()
counts,bin_edges=np.histogram(haberman_died['year'],bins=10,density=True)
pdf=counts/(sum(counts))
print("2.PDF and CDF of Haberman Died on Year of operation")
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('PDF and CDF of Haberman dataset Died on Year of Operation')
plt.xlabel('Year of Operation')
plt.ylabel('Probability')
plt.show()
counts,bin_edges=np.histogram(haberman_died['positive'],bins=10,density=True)
pdf=counts/(sum(counts))
print("3.PDF and CDF of Haberman Died on Positive Auxilary Nodes")
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.title('PDF and CDF of Haberman dataset Died on Positive Auxilary Nodes')
plt.xlabel('Positive Auxilary Nodes')
plt.ylabel('Probability')
plt.show()
```

8.By doing PDF and CDF on survivours the PDF distribution is uniform for age and year of operation but we have most chances of surviving when there are less poitive auxilary nodes.

9.By doing PDF and CDF on the dead we have also got that there are most probable people died who have less positive auxilary nodes.

10.So we have conflict or incoclusive result by PDF and CDF Further Analysis is Required.
